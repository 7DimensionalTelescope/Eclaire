reduction.ipynbで使用するFITSをGoogle Drive上にダウンロードします。

※ My Drive/Colab Notebooks にtestというディレクトリを作成し、13枚のFITS(約30MB)をそこに保存します。
当該ディレクトリは3つ目のコードブロックで指定しているので、適宜設定して下さい。

FITSの内訳

・raw_[00-09].fits : 一次処理の対象となる画像

・dark.fits : ダークフレーム

・flat.fits : フラットフレーム

・bpmask.fits : バッドピクセルマスク

```
from google.colab import drive
drive.mount('/content/drive')
```

```
import shutil
import requests
import os
from tqdm import tqdm_notebook as tqdm
```

```
dirname = '/content/drive/My Drive/Colab Notebooks/test'
if not os.path.exists(dirname):
  os.mkdir(dirname)
 
os.chdir(dirname)
print(os.getcwd())
```

```
url = 'http://sncwall.hp.phys.titech.ac.jp:2388/samplefits/'
fitslist = ['raw%02d.fits'%i for i in range(10)] + ['dark.fits', 'flat.fits', 'bpmask.fits']
```

```
for f in tqdm(fitslist):
  res = requests.get(url+f, stream=True)
  with open(f, 'wb') as fp:
    shutil.copyfileobj(res.raw, fp)
```
